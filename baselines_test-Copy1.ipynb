```python
from RLSwitch import RLSwitchEnv
from stable_baselines3 import A2C,DQN,PPO
from stable_baselines3.common.env_checker import check_env
import numpy as np
```

```python
import matplotlib.pyplot as plt
from tqdm.notebook import tqdm
from maxWeight import maxWeight
```

```python
rho=0.95
```

```python
n = 4
end_t = 10000
env =  RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
```

```python
ppoAgent = PPO('MlpPolicy',env,verbose=1)
ppoAgent.learn(total_timesteps=100000)
```

```python
n = 4
end_t = 1000
env =  RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
dqnAgent = DQN('MlpPolicy',env,
               buffer_size= 100,
               verbose=1)
dqnAgent.learn(total_timesteps=1000000)
```

```python
n = 4
end_t = 1000
env =  RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
a2cAgent = A2C('MlpPolicy',  env,n_steps=1, verbose=1,learning_rate= 0.001)
a2cAgent.learn(total_timesteps=1000000)
```

```python
# Random Environment
n = 4 # nxn
end_t = 1000 # number of time steps to end an episode after
env = RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
# test settings

tests = 10
rewardsRandom = np.zeros((tests,end_t))


for i in tqdm(range(0,tests)):
    env.reset()
    for j in range(0,end_t):
        action = env.action_space.sample()
        observation, reward, done, info = env.step(action)
        rewardsRandom[i][j] = reward
        if done:
            env.reset()
```

```python
# maxWeight Environment
n = 4 # nxn
alpha= 0.1
end_t = 1000 # number of time steps to end an episode after
env = RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
# test settings

tests = 10
rewardsMaxWeight = np.zeros((tests,end_t))

maxWeightAgent = maxWeight(env)

for i in tqdm(range(0,tests)):
    env.reset()
    for j in range(0,end_t):
        action = maxWeightAgent.act(alpha)
        observation, reward, done, info = env.step(action)
        rewardsMaxWeight[i][j] = reward
        if done:
            env.reset()
```

```python
# a2c Environment
n = 4 # nxn
end_t = 1000 # number of time steps to end an episode after
env = RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
# test settings

tests = 10
rewardsa2c = np.zeros((tests,end_t))

for i in tqdm(range(0,tests)):
    observation = env.reset()
    for j in range(0,end_t):
        action,_ = a2cAgent.predict(observation)
        observation, reward, done, info = env.step(action)
        rewardsa2c[i][j] = reward
        if done:
            env.reset()
```

```python
# dqn Environment
n = 4 # nxn
end_t = 1000 # number of time steps to end an episode after
env = RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
# test settings

tests = 10
rewardsDQN = np.zeros((tests,end_t))

for i in tqdm(range(0,tests)):
    observation = env.reset()
    for j in range(0,end_t):
        action,_ = dqnAgent.predict(observation)
        if hasattr(action, "__len__"):
            action = np.random.choice(action)
        observation, reward, done, info = env.step(action)
        rewardsDQN[i][j] = reward
        if done:
            env.reset()
```

```python
# ppo Environment
n = 4 # nxn
end_t = 1000 # number of time steps to end an episode after
env = RLSwitchEnv(n,end_t, lambdaMatrix = None,rho=rho)
# test settings

tests = 10
rewardsPPO = np.zeros((tests,end_t))

for i in tqdm(range(0,tests)):
    observation = env.reset()
    for j in range(0,end_t):
        action,_ = ppoAgent.predict(observation)
        if hasattr(action, "__len__"):
            action = np.random.choice(action)
        observation, reward, done, info = env.step(action)
        rewardsPPO[i][j] = reward
        if done:
            env.reset()
```

```python
# avgRandom = np.average(rewardsRandom,axis=1)
avgDQN = np.average(rewardsDQN,axis=1)
avga2c = np.average(rewardsa2c,axis=1)
avgPPO = np.average(rewardsPPO,axis=1)
avgmaxWeight = np.average(rewardsMaxWeight,axis=1)
```

```python
plt.plot(avgDQN)
plt.plot(avga2c)
plt.plot(avgPPO)
plt.plot(avgmaxWeight)
# plt.plot(avgRandom)
plt.legend([
            "DQN",
            "a2c",
            "avgPPO",
            'maxWeight'
#             ,'Random'
           ]
          )
plt.title("rho= " + str(rho) + " Simulated Rounds vs Average Loss")
plt.xlabel("Simulated Round")
plt.ylabel("Average Loss")
plt.show()
# plt.savefig("graphs/rho "+str(rho)+ "noa2cdqn.png")
```
